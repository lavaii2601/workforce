# [dbo].[ToTicks]

```python
/****** Object:  UserDefinedFunction [dbo].[ToTicks]    Script Date: 3/27/2026 2:19:45 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- =============================================
-- Author:		<Author,,Name>
-- Create date: <Create Date,,>
-- Description:	<Description,,>
-- =============================================
CREATE FUNCTION [dbo].[ToTicks] ( @DateTime datetime2 )
  RETURNS bigint
AS
BEGIN
    DECLARE @Days bigint = DATEDIFF( DAY, '00010101', cast( @DateTime as date ) );
    DECLARE @Seconds bigint = DATEDIFF( SECOND, '00:00', cast( @DateTime as time( 7 ) ) );
    DECLARE @Nanoseconds bigint = DATEPART( NANOSECOND, @DateTime );
    RETURN  @Days * 864000000000 + @Seconds * 10000000 + @Nanoseconds / 100;
END

GO
```

# [dbo].[DoitsuAddBalanceByAccountId]

```python
/****** Object:  StoredProcedure [dbo].[DoitsuAddBalanceByAccountId]    Script Date: 3/27/2026 2:19:46 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[DoitsuAddBalanceByAccountId] @amount float, @membershipCode nvarchar(250)
AS
UPDATE Account SET Balance += @amount WHERE AccountID = (SELECT a.AccountID 
FROM MembershipCard as mc
INNER JOIN Account AS a ON a.MembershipCardId = mc.Id
WHERE mc.MembershipCardCode like N'%'+@membershipCode+'%' and a.[Type] = 1)
GO
```

# [dbo].[PromoMobileCardTo100CardNew]

```python
/****** Object:  StoredProcedure [dbo].[PromoMobileCardTo100CardNew]    Script Date: 3/27/2026 2:19:46 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- =============================================
-- Author:		<Author,,Name>
-- Create date: <Create Date,,>
-- Description:	<Description,,>
-- =============================================
CREATE PROCEDURE [dbo].[PromoMobileCardTo100CardNew]
	-- Add the parameters for the stored procedure here
	@MobileCardCode NVARCHAR(50),
	@PhysicalCardCode NVARCHAR(50)
AS
BEGIN
	DECLARE @MobileCustomerId AS INT
	DECLARE @PhysicalCustomerId AS INT
	DECLARE @MobileCardId AS INT
	DECLARE @PhysicalCardId AS INT

	SET @MobileCustomerId = (SELECT CustomerId FROM MembershipCard WHERE MembershipCardCode = @MobileCardCode)
	SET @PhysicalCustomerId = (SELECT CustomerId FROM MembershipCard WHERE MembershipCardCode = @PhysicalCardCode)
	SET @MobileCardId = (SELECT Id FROM MembershipCard WHERE MembershipCardCode = @MobileCardCode)
	SET @PhysicalCardId = (SELECT Id FROM MembershipCard WHERE MembershipCardCode = @PhysicalCardCode)
	
	BEGIN TRANSACTION;  
	
	BEGIN TRY  
    INSERT INTO Account
	SELECT
      CAST([AccountCode] as bigint) + 1
      ,[AccountName]
      ,[Level_]
      ,[StartDate]
      ,[FinishDate]
      ,[Balance]
      ,[ProductCode]
      ,@PhysicalCardId
      ,[Type]
      ,[BrandId]
      ,[Active]
	FROM [passio].[dbo].[Account] Where MembershipCardId = @MobileCardId and [Type] = 1

	UPDATE Account SET Balance = 0 Where MembershipCardId = @MobileCardId and [Type] = 1
	
	UPDATE Customer 
		Set Phone = (Select Phone From Customer Where CustomerId = @MobileCustomerId),
		AccountPhone = (Select AccountPhone From Customer Where CustomerId = @MobileCustomerId),
		FacebookId = (Select FacebookId From Customer Where CustomerId = @MobileCustomerId),
		Email = (Select Email From Customer Where CustomerId = @MobileCustomerId),
		Name = (Select Name From Customer Where CustomerId = @MobileCustomerId),
		picURL = (Select picURL From Customer Where CustomerId = @MobileCustomerId)
	Where CustomerID = @PhysicalCustomerId

	UPDATE MembershipCard SET CustomerId = @PhysicalCustomerId, [Status] = 1, PhysicalCardCode = @PhysicalCardCode Where Id in (@PhysicalCardId,@MobileCardId)
	END TRY  
	BEGIN CATCH  
    IF @@TRANCOUNT > 0  
        ROLLBACK TRANSACTION;  
	END CATCH;  

	IF @@TRANCOUNT > 0  
    COMMIT TRANSACTION;   
END

GO
```
